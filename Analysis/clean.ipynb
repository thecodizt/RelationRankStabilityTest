```python
def parse_response(response):
    response = response.strip()
    response = response.split(' ')

    for i in range(len(response) - 1, -1, -1):
        try:
            # remove trailing symbols before parsing

            response[i] = response[i].strip()
            response[i] = response[i].strip('.')
            response[i] = response[i].strip(',')
            response[i] = response[i].strip('!')
            response[i] = response[i].strip('?')
            response[i] = response[i].strip(':')
            response[i] = response[i].strip(';')
            
            response[i] = float(response[i])
            if response[i] >= 0 and response[i] <= 1:
                return response[i]
        except:
            pass

    return -1
```

```python
results['clean_result'] = results['response'].apply(parse_response)
```
